# Comparación con la solución real

La solución real de la ecuación diferencial es: $y(t)=(t+1)^2-0.5$

```python
%load_ext autoreload
```

```python
from src import ODE_euler

f = lambda t, y: y - (t)**2 + 1
a = 0
b = 2
y_t0 = 0.5
N = 10 

# Método de Euler
ys, ts, h = ODE_euler(a=a, b=b, f=f, y_t0=y_t0, N=N)
```

```python
# Solución real
def real_solution(t):
    return (t + 1)**2 - 0.5
```

```python
import numpy as np
# Valores para la solución real
t_real = np.linspace(a, b, 10)
y_real = real_solution(t_real)

# Imprimir valores obtenidos por el método de Euler
print("Valores obtenidos por el método de Euler:")
for t, y in zip(ts, ys):
    print(f"t: {t:.2f}, y: {y:.4f}")
```

```python
import matplotlib.pyplot as plt
# Graficar
plt.figure(figsize=(6, 4))
plt.plot(ts, ys, 'o-', label='Método de Euler', color='blue')
plt.plot(t_real, y_real, label='Solución Real', color='orange')
plt.title('Solución Real de Ejercicio1')
plt.xlabel('Tiempo')
plt.ylabel('y')
plt.legend()
plt.grid()
plt.show()
```
